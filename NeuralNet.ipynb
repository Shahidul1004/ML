```python
import numpy as np
import pandas as pd
import random as random
from layers import*
from svm import*
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot
from NeuralNet import NeuralNetwork
from keras.utils import to_categorical
from keras.datasets import cifar10
(X_train, Y_train), (X_test, Y_test) = cifar10.load_data()
```

```python
X_train=X_train.reshape(X_train.shape[0], -1)
X_test=X_test.reshape(X_test.shape[0], -1)
Y_train=Y_train.reshape(Y_train.shape[0])
Y_test=Y_test.reshape(Y_test.shape[0])
```

```python
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
num_feature=X_train.shape[1]
num_class=len(np.unique(Y_train))
num_train=X_train.shape[0]
num_test=X_test.shape[0]
```

```python
hidden_layer=[1000, 1000]
steps=200
reg=.001
l_rate=0.01
batchnorm=1
dropout=0
net=NeuralNetwork(num_train, num_feature, hidden_layer, num_class, steps, reg, l_rate, batchnorm, dropout)
loss, grads, acc=net.train(X_train, Y_train, 5000, X_test, Y_test)
#accuracy=net.predict(X_test, Y_test)
#print(accuracy/num_test)
plt.subplot(1, 2, 1)
plt.plot(loss)
plt.subplot(1, 2, 2)
plt.plot(acc)
```
